## A function was created to insert the missing values directly

```python
def replace_missing_data (male, female,country,gender_pop21):
    gender_pop21.set_index(['Country Name'],inplace = True)
    for i in gender_pop21['Indicator Name']:
        if i == "male":
            gender_pop21.at[country, "2021"] = male
        elif i == "female":
            gender_pop21.at[country, "2021"] = female
    gender_pop21 = gender_pop21.reset_index()
    return gender_pop21
```

# To Conduct Further Analysis on Dataset for Statistical Changes for five(5) years

```python
gender_pop.isnull().sum()
```

```python
clean_gender_pop = gender_pop.dropna(subset = ['1960'])
clean_gender_pop = clean_gender_pop.dropna(subset = ['2021'])
clean_gender_pop.isnull().sum()
```

```python
full_gender_pop5y = clean_gender_pop.drop(clean_gender_pop.loc[:, "1960":"2016"], axis = 1)
full_gender_pop5y
```

```python
gender_pop5y = full_gender_pop5y.drop(['Country Name', 'Country Code'], axis = 1)
gender_pop5y
```

```python
gender_pop5y = gender_pop5y.groupby(['Indicator Name']).sum()
gender_pop5y = gender_pop5y
gender_pop5y
```

```python
gender_pop5y = gender_pop5y.transpose()
gender_pop5y
```

```python
#horizontal = gender_pop5y['2017','2018','2019','2020','2021']
gender_pop5y.plot(label = 'Indicator Name', color = ('hotpink','skyblue'), marker = "p")
plt.title('Five Years Gender Global Population')
plt.xlabel('Years')
plt.legend()
```
